```python
#This project is about read data from dataset, Perform statistical analysis (Mean, Median, Mode Range, Interquartile range (IQR),
#1st and 3rd quartiles, Sample variance, and sample standard deviation), Find Outliers and present results in graphical format.

#Importing necessary libraries

import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read the data from csv file of dataset,, encoding (cp1252), and low_memory=False to handle any memory allocation issues.
df = pd.read_csv('new_file.csv', encoding='cp1252', low_memory=False)
```

```python
# Convert 'square_feet' and 'price_display' columns to numeric and drop rows with missing values
df['square_feet'] = pd.to_numeric(df['square_feet'], errors='coerce')
df['price_display'] = pd.to_numeric(df['price_display'], errors='coerce')
df.dropna(subset=['square_feet', 'price_display'], inplace=True)
```

```python
#To show the dataset table 
df
```

Perform statistical analysis

```python
#Calucating Mean for price_display
price_mean = df['price_display'].mean()

# Rounding the mean value to 2 decimal points
rounded_price_mean = round(price_mean, 2)

print("Mean of price_display:", rounded_price_mean)
```

```python
# Calculating Mean for square_feet
square_feet_mean = df['square_feet'].mean()

# Rounding the mean value to 2 decimal points
rounded_square_feet_mean = round(square_feet_mean, 2)

print("Mean of square_feet:", rounded_square_feet_mean)
```

```python
# Calculate Median for price_display
price_median = df['price_display'].median()

# Round the value up to 2 decimal points
rounded_price_median = round(price_median, 2)

print("Median of price display:", rounded_median)
```

```python
# Calculating Median for square_feet
square_feet = df['square_feet'].median()

# Round the value up to 2 decimal points
rounded_square_feet = round(square_feet, 2)

print("Median of square_feet:", rounded_square_feet)
```

```python
#Calucating Mode for price_display
price_mode = df['price_display'].mode().iloc[0]
print("Mode of price display:",price_mode)
```

```python
# Calculate Mode for square_feet
square_feet = df['square_feet'].mode().iloc[0]
print("Mode of square_feet:",square_feet)
```

```python
#Calucating range of Minimum and Maxium value for 'price_display'
price_display_range = df['price_display'].max() - df['price_display'].min()
print("Range of price display:", price_display_range)
```

```python
# Calculate range of Minimum and Maxium value for 'square_feet'
square_feet_range = df['square_feet'].max() - df['square_feet'].min()
print("Range of square feet:", square_feet_range)
```

```python
# Calculate sample variance for 'price_display'
price_display_variance = df['price_display'].var()

# Round the value up to 2 decimal points
rounded_price_display_variance = round(price_display_variance, 2)

print("Sample variance of price display:", rounded_price_display_variance )
```

```python
# Calculate sample variance for 'square_feet'
square_feet_variance = df['square_feet'].var()

# Round the value up to 2 decimal points
rounded_square_feet_variance = round(square_feet_variance, 2)

print("Sample variance of square feet:", rounded_square_feet_variance)
```

```python
# Calculate sample standard deviation for 'price_display'
price_display_std = df['price_display'].std()
rounded_price_display_std = round(price_display_std, 2)
print("Sample standard deviation of price display:", rounded_price_display_std)
```

```python
# Calculate sample standard deviation for 'square_feet'
square_feet_std = df['square_feet'].std()
rounded_square_feet_std = round(square_feet_std, 2)
print("Sample standard deviation of square feet:", rounded_square_feet_std)
```

```python
# Calculate quartiles (Q1 and Q3) for 'price_display'

price_display_q1 = df['price_display'].quantile(0.25)
print("Q1 of price display:", price_display_q1)

price_display_q3 = df['price_display'].quantile(0.75)
print("Q3 of price display:", price_display_q3)
```

```python
# Calculate quartiles (Q1 and Q3) for 'square_feet'

square_feet_q1 = df['square_feet'].quantile(0.25)
print("Q1 of square feet:", square_feet_q1)

square_feet_q3 = df['square_feet'].quantile(0.75)
print("Q3 of square feet:", square_feet_q3)
```

```python
# Calculate interquartile range (IQR) for 'price_display'
price_display_iqr = price_display_q3 - price_display_q1
print("IQR of price display:", price_display_iqr)
```

```python
# Calculate interquartile range (IQR) for 'square_feet'
square_feet_iqr = square_feet_q3 - square_feet_q1
print("IQR of square feet:", square_feet_iqr)
```

```python
# Calculate lower and upper limits for 'price_display'

price_display_lower_limit = price_display_q1 - 1.5 * price_display_iqr
print("Lower limit of price display:", price_display_lower_limit)

price_display_upper_limit = price_display_q3 + 1.5 * price_display_iqr
print("Upper limit of price display:", price_display_upper_limit)
```

```python
# Calculate lower and upper limits for 'square_feet'

square_feet_lower_limit = square_feet_q1 - 1.5 * square_feet_iqr
print("Lower limit of square feet:", square_feet_lower_limit)

square_feet_upper_limit = square_feet_q3 + 1.5 * square_feet_iqr
print("Upper limit of square feet:", square_feet_upper_limit)
```

```python
# Calculate z-scores for 'price_display'
z_scores = np.abs((df['price_display'] - df['price_display'].mean()) / df['price_display'].std())

# Define the threshold for identifying outliers
threshold = 3

# Identify outliers
outliers = df[z_scores > threshold]

# Print the outliers
print("Outliers:")
print(outliers)
```

```python
# Calculate z-scores for 'price_display'
z_scores = np.abs((df['square_feet'] - df['square_feet'].mean()) / df['square_feet'].std())

# Define the threshold for identifying outliers
threshold = 3

# Identify outliers
outliers = df[z_scores > threshold]

# Print the outliers
print("Outliers:")
print(outliers)
```

```python
# Create a box plot for 'price_display'
plt.figure(figsize=(20,10))
sns.boxplot(x=df['price_display'])
plt.xlabel('price_display', fontsize=25)
plt.title('Box Plot - price_display', fontsize=25)
plt.tick_params(axis='x', labelsize=20)
plt.show()

#The plot represents the box plot for the 'price_display' column. The box in the plot represents the interquartile range (IQR), which contains the middle 50% of the data. The line inside the box indicates the median (50th percentile) of the data. The whiskers extend to the minimum and maximum values within 1.5 times the IQR. Any points outside the whiskers are considered outliers and plotted individually. The distribution of 'price_display' can be analyzed based on the position of the median, the spread of the box (IQR), and the presence of outliers.
```

```python
# Create a box plot for 'square_feet'
plt.figure(figsize=(20,10))
sns.boxplot(x=df['square_feet'])
plt.xlabel('square_feet', fontsize=25)
plt.title('Box Plot - square_feet', fontsize=25)
plt.tick_params(axis='x', labelsize=20)
plt.show()
```

```python
# Create a scatter plot using Seaborn for 'square_feet' and 'price_display'
sns.scatterplot(x='square_feet', y='price_display', data=df)
plt.xlabel('square_feet')
plt.ylabel('price_display')
plt.title('Seaborn Scatter Plot - square_feet vs price_display')
plt.show()
```

```python
# Create a linear regression plot using Seaborn for 'square_feet' and 'price_display'
sns.regplot(x='square_feet', y='price_display', data=df)
plt.xlabel('square_feet')
plt.ylabel('price_display')
plt.title('Linear Regression - square_feet vs price_display')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['square_feet', 'price_display']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```
